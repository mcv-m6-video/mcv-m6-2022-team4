```python
camera_imgs = {}

for camera in cameras:
    
    frames =[]
    print (f'processing camera {camera}')
    
    cam_files = sorted([f for f in files if camera in f])
    print (len(cam_files))
    print (cam_files[:5], cam_files[0])
    
    for f in cam_files:
        sample = cv2.imread(f)
        sample = cv2.cvtColor(sample, cv2.COLOR_BGR2RGB)
        sample = cv2.resize(sample, (1280,720), interpolation = cv2.INTER_AREA)
        frames.append(sample)
        
        
    print (len(frames))
    print (sample.shape)
    plt.imshow(sample)
    plt.show()

    camera_imgs[camera] = frames[4:] #np.array(frames)
    print (20*'....'); print()
```

```python
for camera, frames in camera_imgs.items():
    print (camera, len(frames))
    imageio.mimsave(f'scene3_{camera}.gif', frames)
```
